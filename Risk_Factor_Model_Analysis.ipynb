```python
AAPL_df = pdr.data.get_data_yahoo('AAPL', start=start)['Adj Close'].resample('M').ffill().pct_change()
```

```python
AAPL_df= pd.DataFrame(AAPL_df)
AAPL_df
```

```python
AAPL_df.index.dtype
```

```python
df_merged.index.dtype
```

```python
## Changing the type of index for Apple stock
AAPL_df['str_date'] = AAPL_df.index.astype(str)
AAPL_df['dt_date'] =pd.to_datetime(AAPL_df['str_date']).dt.strftime('%Y-%m')
```

```python
AAPL_df.dt_date.dtype
```

```python
## Changing the type of index for Apple stock
df_merged['str_date'] = df_merged.index.astype(str)
df_merged['dt_date'] =pd.to_datetime(df_merged['str_date']).dt.strftime('%Y-%m')
```

```python
df_merged.dt_date.dtype
```

```python
AAPL_df_merged = pd.merge(AAPL_df, df_merged,how='inner',on='dt_date',sort=True,copy=True,indicator=False,validate='one_to_one')
```

```python
AAPL_df_merged
```

```python
AAPL_df_merged.drop(columns=['str_date_x','str_date_y'],inplace=True)
```

```python
AAPL_df_merged.rename(columns={'Adj Close':'AAPL'},inplace=True)
```

```python
AAPL_df_merged
```

```python
AAPL_df_merged['AAPL_RF'] =AAPL_df_merged['AAPL']*100-AAPL_df_merged['RF']
```

```python
AAPL_df_merged
```

```python
AAPL_df_merged.dropna(axis=0,inplace=True)
```

```python
AAPL_df_merged.rename(columns={'Mom   ':'MOM'}, inplace=True)
```

```python
from statsmodels.api import OLS
result = OLS(AAPL_df_merged['AAPL_RF'],AAPL_df_merged[['Mkt-RF','SMB','HML','MOM']],missing='drop').fit()
```

```python
result.summary()
```

```python
import statsmodels.tools
AAPL_df_merged_c =statsmodels.tools.add_constant(AAPL_df_merged,prepend=True)
```

```python
AAPL_df_merged_c
```

```python
result = OLS(AAPL_df_merged_c['AAPL_RF'],AAPL_df_merged_c[['const','Mkt-RF','SMB','HML','MOM']],missing='drop').fit()
result.summary()
```
